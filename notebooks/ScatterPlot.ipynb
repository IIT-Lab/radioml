```python
import commpy as cp
import numpy as np

import pylab
import matplotlib.pyplot as plt
```

```python
qam16 = cp.modulation.QAMModem(m=4)

# Generate 1000 noisy symbols
msg_bits  = np.random.uniform(0, 2, 300).astype(int)
modulated = qam16.modulate(msg_bits)
noisy     = cp.channels.awgn(modulated, snr_dB=10)

# Colormap is the ground truths (original symbols)
mapping, ground_truths = np.unique(modulated, return_inverse=True)

# Scatter plot
plt.scatter(x=np.real(noisy), 
            y=np.imag(noisy), 
            c=ground_truths)
plt.scatter(x=np.real(qam16.constellation), 
            y=np.imag(qam16.constellation), 
            c='green')

# Show the line at 0 origin
plt.axvline(c='black')
plt.axhline(c='black')
```
